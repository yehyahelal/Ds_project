# Machine learning

Applying machine learning to our project as a concept was to be able to make predictions based on trained data that is provided, starting off by importing the scikit library to use linear regression, then creating an instance/object from the class linear regression, afterwards we declare 2 variables, one of them to hold the values that will be used to train and make the predictions based off, and the other will hold the predicted value, in our case we are trying to predict the price of a car based on the year, id and kilometres given. Tests are made with random values and a prediction is made based on those values. We train, test and split the data then fit the trained data inside the linear regression model. Afterwards we use the pred method to predict the price of a car with given values that are listed as mentioned previously.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
LRmodel= LinearRegression(n_jobs=-1)
X = data[['Year','Kilometres']]
y = data['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =2, random_state=0)
LRmodel.fit(X_train, y_train)
```

```python
y_pred = LRmodel.predict(X_test)
print("Prediction:", y_pred)
```
